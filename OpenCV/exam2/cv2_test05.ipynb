```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros(shape = (512, 512, 3), dtype=np.uint8) +255
```

```python
pt1 = 100, 100
pt2 = 400, 400
cv2.rectangle(img, pt1, pt2, (0, 255, 0), 2)
print(0)
```

```python
ptCenter = 250, 250
size = 300, 300 
box = (ptCenter, size, 0)
#cv2.ellipse(img, box, (0,0,255), 1)

cv2.circle(img, ptCenter, radius = 150, color = (255, 0, 0), thickness = 1)
print(0)
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
cv2.line(img, pt1, pt2, (255, 0, 0), 2)
cv2.line(img, (400, 100), (100, 400), (255, 0, 0), 2)

print(0)
```
