```python
import cv2
import matplotlib.pyplot as plt
```

```python
imageFile = './opencv_data/lena.jpg'
```

```python
img = cv2.imread(imageFile)
img3 = cv2.imread(imageFile, 0)
```

```python
cv2.imshow('lena', img3)
cv2.waitKey(5000) # opencv 영상 멈춤
cv2.destroyAllWindows() # 윈도운 메시지로 창을 종료
```

```python
cv2.imshow('lena color', img)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
plt.show()
```
